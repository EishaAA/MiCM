# Part 4: Recursion

**Recursion** is a problem solving technique where a problem is broken down into smaller intances of the same problem.

A **recursive function** is a function that calls itself during execution.

Suppose we wanted to compute the factorial of a number. Recall that $n! = 1*2*\dots*(n-1)*n$.

One way we could compute this by looping through and multiplying all the numbers 1 to $n$. However, we also notice that $n! = n*(n-1)!$ - in other words, a factorial can be broken down into another, smaller factorial problem. Let's see how we would implement this in Python:

```
def factorial1(n):
    """Solved using loops"""
    result = 1
    if n > 0:
        for i in range(1, n+1):
            result *= i
    return result

def factorial2(n):
    """Solved using recursion"""
    if n == 0:
        return 1
    else:
        return n*factorial2(n-1)

print(factorial1(5))
print(factorial2(5))
```

Let's look at another example. We can implement the Fibonacci sequence using recursion as well.

```
def fibonacci(n):
    """Print the first nth terms of the Fibonacci sequence"""
    if n <= 1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))

#  Let's print the first 10 terms of the sequence
for i in range(5):
    print(fibonacci(i))
```
